```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./ecommerce_sales.csv')

df = pd.DataFrame(data)
df.head()
```

```python
df.drop(['competition_intensity','inventory_pressure'],axis=1,inplace=True)
df.head()
```

```python
plt.bar(df['order_date'],df['revenue'])
plt.xlabel('Dates')
plt.ylabel('Revenue')
plt.title('Bar Plot')
plt.show()
```

```python
zones = df.groupby('zone')['units_sold'].sum()

plt.pie(zones,labels=zones.index,autopct='%1.1f%%')
plt.title("Zone wise Sales")
plt.show()
```
